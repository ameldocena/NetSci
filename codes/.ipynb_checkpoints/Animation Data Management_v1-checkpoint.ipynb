#### Ensuring the values of each variable is unique for each film data

```python
def print_full(x):
    pd.set_option('display.max_rows', len(x))
    print(x)
    pd.reset_option('display.max_rows')
```

```python
def verify_unique(df, var):
    df['ones'] = 1
    unique = df.groupby(var)['ones'].sum()
    print("Verify: {}. Total entries: {}. Unique entries: {}".format(var, len(df.index), len(unique)))
    print_full(df.groupby(var)['ones'].sum())
```

```python
verify_unique(data_pixar, 'crew')
```

#### Create fully connected graph among the nodes for each film

##### Model 1
In this model we shall consider only the directors, story writers, screenplay and producers. And we presume that they are knitted together.

```python
#Check the unique values of dept
data_pixar['dept'].unique()

#Clean up the values and store the extracted in data_pixar1
data_pixar1 = data_pixar.replace({'dept': {'director': 'direction', 'co-director': 'direction'}})
print("unique values dept", data_pixar1['dept'].unique())

#Extract values
data_pixar1.loc['dept'] = 'direction'


#Drop duplicates. Each node should be unique
#data_pixar1 = data_pixar1.drop_duplicates(['a','b'])
```

```python
#Extract values
scope = ['direction', 'writers', 'screenplay', 'producer', 'art direction']
model1 = data_pixar1.loc[data_pixar1['dept'].isin(scope)].loc[data_pixar1['film']=='ToyStory']
crew1 = model1.drop_duplicates('crew')

model2 = data_pixar1.loc[data_pixar1['dept'].isin(scope)].loc[data_pixar1['film']=='Cars']
crew2 = model2.drop_duplicates('crew')
print_full(crew1)
print_full(crew2)
```

```python
import itertools as it
nodes1 = crew1['crew']
edges1 = pd.DataFrame(it.combinations(nodes1, 2))
edges1['film'] = 'ToyStory'

nodes2 = crew2['crew']
edges2 = pd.DataFrame(it.combinations(nodes2, 2))
edges2['film'] = 'Cars'

edges = edges1.append(edges2).rename(columns = {'0': 'source', '1': 'target'})
print_full(edges)
```

```python
G = nx.from_pandas_edgelist(edges, source = 0, target = 1, edge_attr=None, create_using=nx.Graph())
```

```python
dir(nx)
```

```python
G_ex = nx.Graph()
G_ex.add_nodes_from(nodes1)
print("\nNodes 1:", G_ex.nodes)
G_ex.add_nodes_from(nodes2)
print("\nNodes 1 & 2:", G_ex.nodes)

G_ex.add_edges_from(edges1)
print("\nEdges 1: {}, {}".format(len(G_ex.edges), G_ex.edges))

G_ex.add_edges_from(edges2)
print("\nEdges 2: {}, {}".format(len(G_ex.edges), G_ex.edges))
```

```python
G_ex.edges()
```

```python
plt.figure(figsize = (20, 20))
nx.draw_spring(G_ex, with_labels=True)
```
